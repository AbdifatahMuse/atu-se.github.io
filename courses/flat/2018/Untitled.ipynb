```python
from automata.fa.nfa import NFA
from automata.fa.dfa import DFA
from graphviz import Digraph
```

Now here is our general function for printing a finite automaton.

```python
def generate_fa_image(dfa, name="Finite Automaton"):
    dot = Digraph(name)
    for i in dfa.states:
        if(i in dfa.final_states):
            dot.attr('node', shape='doublecircle')
        else:
            dot.attr('node', shape='circle')
        dot.node(i)

    for start_state, values in dfa.transitions.items():
        for symbol, next_state in values.items():
            dot.edge(start_state,next_state, symbol)

    dot.attr('node', shape="point")
    dot.node("start_point_xyza")
    dot.edge("start_point_xyza", dfa.initial_state) 
    return dot
```

```python
dfa1 = DFA(
   ...:     states={'q0', 'q1', 'q2'},
   ...:     input_symbols={'0', '1'},
   ...:     transitions={
   ...:         'q0': {'0': 'q0', '1': 'q1'},
   ...:         'q1': {'0': 'q0', '1': 'q2'},
   ...:         'q2': {'0': 'q2', '1': 'q1'}
   ...:     },
   ...:     initial_state='q0',
   ...:     final_states={'q1'}
   ...: )

image =generate_fa_image(dfa1, "DFA 1")
image
```

```python
dfa2 = DFA(
    states = {'A', 'B', 'C', 'D', 'E'},
    input_symbols = {'a', 'b'},
    initial_state='A',
    final_states={'C', 'D', 'E'},
    transitions = {
        'A': {'a':'B', 'b':'A'},
        'B': {'a':'C', 'b':'B'},
        'C': {'a':'D', 'b':'E'},
        'D': {'a':'D', 'b':'D'},
        'E': {'a':'D', 'b':'D'}
    }
)

image = generate_fa_image(dfa2, "DFA 2")
image
```
